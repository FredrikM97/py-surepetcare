```python
import os
from pathlib import Path
from dotenv import load_dotenv
from surepcio import SurePetcareClient, Household

load_dotenv(Path(".env"))

# Read credentials
email = os.getenv("EMAIL")
password = os.getenv("PASS")


client = SurePetcareClient()
await client.login(email=email, password=password)
households = await client.api(Household.get_households())

(hub, feeder) = await client.api(Household({"id": households[0].id}).get_devices())

(lilo, maui, bricka) = await client.api(Household({"id": households[0].id}).get_pets())

await client.api(maui.refresh())
```

```python
from surepcio.enums import PetDeviceLocationProfile

await client.api(maui.set_profile(feeder.id, PetDeviceLocationProfile.INDOOR_ONLY))
```

```python
from surepcio.command import Command
from surepcio.const import API_ENDPOINT_PRODUCTION
from surepcio.enums import PetDeviceLocationProfile


await client.api(
    Command(
        method="PUT",
        endpoint=f"{API_ENDPOINT_PRODUCTION}/device/1376800/tag/{maui.tag}/async",
        params={"profile": PetDeviceLocationProfile.INDOOR_ONLY.value},
        device=maui,
    )
)
```

```python
from surepcio.command import Command
from surepcio.const import API_ENDPOINT_PRODUCTION

await client.api(
    Command(
        method="GET",
        endpoint=f"{API_ENDPOINT_PRODUCTION}/household/{maui.household_id}/device/control/status",
        params={"status": [5, 1, 2, 3, 4]},
        full_response=True,
    )
)
```
